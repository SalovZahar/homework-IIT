##<font size="5" color="black" face="Times"><b>&emsp;3.1. Разбор теоретических вопросов</b></font><a name = 'part31'>

<font size="3" color="black" face="Times"><div align="justify">
<b>&emsp;Теория по HTML, теги, используемые при выполнении практической работы</b></div></font>

<font size="3" color="black" face="Times"><div align="justify">
&emsp;Web - Распределённая система, предоставляющая доступ к связанным между собой документам, расположенным на различных компьютерах, подключённых к сети Интернет.
Всемирную паутину образуют сотни миллионов веб-серверов.

&emsp;HTML расшифровывается как HyperText Markup Language (язык гипертекстовой разметки). Это язык разметки документов во Всемирной паутине (World Wide Web, WWW).

&emsp;HTML — это стандартизированный язык, позволяющий составлять форматированный текст. Браузер интерпретирует его и отображает на экране элементы веб-страниц.

&emsp;Тег — это специальное служебное слово, заключенное в угловые скобки. Его ещё называют «элемент HTML».

&emsp;Использование тегов в практической работе:

1. **\<head>** - Элемент-контейнер для метаданных HTML-документа. Например:
   ```bash
   <head>
    <meta charset="UTF-8">
    <title>Главная страница</title>
    <link rel="stylesheet" href="css/styles.css">
    <link rel="icon" href="images/favicon.ico" type="image/x-icon">
</head>
   ```

2. **\<body>** - Представляет тело документа (содержимое, не относящееся к метаданным документа). Например:
   ```bash
   <body>
       <main>
            <h2>О сайте</h2>
            <p>Это главная страница моего сайта. Здесь вы найдете ссылки на другие страницы.</p>
        </main>
   </body>
   ```

3. **\<header>** - Секция для вводной информации сайта или группы навигационных ссылок. Например:
   ```bash
   <header>
       <h1>Добро пожаловать на мой сайт</h1>
   <\header>
   ```

4. **\<main>** - Контейнер для основного уникального содержимого документа. На одной странице должно быть не более одного элемента \<main>. Например:
   ```bash
   <main>
        <h2>О сайте</h2>
        <p>Это главная страница моего сайта. Здесь вы найдете ссылки на другие страницы.</p>
    </main>
   ```

5. **\<aside>** - Представляет контент страницы, который имеет косвенное отношение к основному контенту страницы/сайта. Например:
   ```bash
   <aside>
        <h3>Боковая панель</h3>
        <p>Дополнительная информация может быть размещена здесь.</p>
    </aside>
   ```

6. **\<footer>** - Определяет завершающую область (нижний колонтитул) документа или раздела. Например:
   ```bash
   <p>&copy; 2024 Салов Захар. Все права защищены.</p>
   ```

7. **\<html>** - Корневой элемент HTML-документа. Сообщает браузеру, что это HTML-документ. Является контейнером для всех остальных html-элементов. Например:
   ```bash
   <html>
      ...
    </html>
   ```

8. **\<img>** - Встраивает изображения в HTML-документ с помощью атрибута src, значением которого является адрес встраиваемого изображения. Например:
   ```bash
   <img src="images/photo1.jpg" alt="Фото 1">
   ```

9. **\<audio>** - Загружает звуковой контент на веб-страницу. Например:
   ```bash
   <audio controls>
        <source src="audio/sample.mp3" type="audio/mpeg">
        Ваш браузер не поддерживает аудио элемент.
    </audio>
   ```

10. **\<audio>** - Добавляет на страницу видео-файлы. Поддерживает 3 видео формата: MP4, WebM, Ogg. Например:
   ```bash
   <video controls>
        <video controls width="400">
        <source src="video/sample.mp4" type="video/mp4">
    </video>
   ```


##<font size="5" color="black" face="Times"><b>&emsp;5.4. Задача 4</b></font><a name = 'part54'>

<font size="3" color="black" face="Times"><div align="justify">
Изменяем главную страницу и создаем в ней Заголовок, Описание, Ссылки на все дочерние страницы.
Создаем на каждой из дочерних страниц ссылки для перехода на главную страницу.

1. Используем тег \<title> для создания заголовка страницы
2. Используем теги \<h1>, \<h2>, \<h3> и \<p> для создания описания страницы
3. В теге навигации \<nav> создаем ссылки (тег \<a>) для перехода на все дочерние страницы при помощи непронумерованного списка (тег \<ul> а в нем \<li>)

```
<!DOCTYPE html>
<html lang="ru">
<head>
    <meta charset="UTF-8">
    <title>Главная страница</title>
</head>
<body>
    <header>
        <h1>Добро пожаловать на мой сайт</h1>
    </header>

    <nav>
        <ul>
            <li><a href="index.html">Главная</a></li>
            <li><a href="page1.html">Фотогалерея</a></li>
            <li><a href="page2.html">Аудио</a></li>
            <li><a href="page3.html">Видеогалерея</a></li>
        </ul>
    </nav>

    <main>
        <h2>О сайте</h2>
        <p>Это главная страница моего сайта. Здесь вы найдете ссылки на другие страницы.</p>
    </main>

    <aside>
        <h3>Боковая панель</h3>
        <p>Дополнительная информация может быть размещена здесь.</p>
    </aside>

    <footer>
        <p>&copy; 2024 Салов Захар. Все права защищены.</p>
    </footer>
</body>
</html>
```

В дочерних файлах создаем в пустом теге \<nav> навигационные ссылки (как в главном файле):

```
<nav>
        <ul>
            <li><a href="index.html">Главная</a></li>
            <li><a href="page1.html">Фотогалерея</a></li>
            <li><a href="page2.html">Аудио</a></li>
            <li><a href="page3.html">Видеогалерея</a></li>
        </ul>
    </nav>
```
